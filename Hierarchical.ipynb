```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, precision_score
```

```python
df = pd.read_csv('./Cust_Spend_Data.csv')
df.head()
```

```python
sb.pairplot(df, diag_kind= 'kde');
```

## Dendogram

```python
X = df.drop(['Cust_ID', 'Name'], axis= 1)
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

## Hierarchical (Agglomerative)

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters= 2, affinity='euclidean', linkage= 'ward')
cluster.fit_predict(X)
```

```python
plt.scatter(X.Avg_Mthly_Spend, X.No_Of_Visits, c = cluster.labels_, cmap='rainbow');
```

## Kmeans

```python
from sklearn.cluster import KMeans
cluster2 = KMeans(n_clusters=2)
cluster2.fit_predict(X)
```

```python
plt.scatter(X.Avg_Mthly_Spend, X.No_Of_Visits, c = cluster2.labels_, cmap='rainbow');
```
